```python
import pandas as pd
import matplotlib.pyplot as plt
import numpy as np
#Read data
data= pd.read_csv("data.csv")
```

```python
#Descriptor type
descriptor_type = list(set(data['Descriptor Type']))
# Detector type
detector_type= list(set(data['Detector Type']))
```

```python
def plot_data(descriptor_type, detector_type):
    fig, ax = plt.subplots()
    fig.set_size_inches(18.5, 12.5, forward=True)
    res= data[(data['Descriptor Type']== descriptor_type)  & (data['Detector Type']== detector_type)]
    ax.plot(res['Frame_ID'],res['Camera TTC (s)'], label="camera ttc (sec)")
    ax.plot(res['Frame_ID'],res['Lidar TTC(s)'],label="lidar ttc(sec)")
    ax.plot(res['Frame_ID'], res['Difference TTC'],label="diff ttc (sec)")
    ax.set_xlabel('Frame ID')
    ax.set_xticks(range(1,19))
    ax.set_yticks(np.arange(np.min(res['Difference TTC']), np.max(np.max(res[['Lidar TTC(s)','Camera TTC (s)']])), 1.))
    legend = ax.legend(loc='upper right', shadow=True, fontsize='x-large')
    plt.savefig('plots/'+'desc_'+descriptor_type + '_detect_'+ detector_type+'.png')
```

```python
for detector in detector_type:
    for descpt in descriptor_type:
        try:
            plot_data(descpt,detector)
        except:
            continue
```
